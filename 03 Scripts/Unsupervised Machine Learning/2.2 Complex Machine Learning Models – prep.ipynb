```python
import pandas as pd
import numpy as np

# Load the data
weather_obs = pd.read_csv('/Users/davidscheider/anaconda_projects/ClimateWins/02 Data/Original Data/Dataset-weather-prediction-dataset-processed.csv')
pleasant_weather = pd.read_csv('/Users/davidscheider/anaconda_projects/ClimateWins/02 Data/Original Data/Dataset-Answers-Weather_Prediction_Pleasant_Weather.csv')

print("="*80)
print("IDENTIFICATION OF MISSING VALUES")
print("="*80)

#  Remove GDANSK, ROMA, TOURS
stations_to_remove = ['GDANSK', 'ROMA', 'TOURS']

# Find all columns
cols_to_drop = [col for col in weather_obs.columns if any(station in col for station in stations_to_remove)]
print(f"\n1. Zu entfernende Spalten (Stationen: {stations_to_remove}):")
print(f"   Anzahl: {len(cols_to_drop)}")
print(f"   Spalten: {cols_to_drop}")

# drop coluns 
weather_cleaned = weather_obs.drop(columns=cols_to_drop)

print(f"\n2. Shape after drop of GDANSK, ROMA, TOURS: {weather_cleaned.shape}")

# Check weather stations for missing values
print("\n3. Missing valies per Station:")
print("="*80)

remaining_stations = ['BASEL', 'BELGRADE', 'BUDAPEST', 'DEBILT', 'DUSSELDORF', 
                      'HEATHROW', 'KASSEL', 'LJUBLJANA', 'MAASTRICHT', 'MADRID', 
                      'MUNCHENB', 'OSLO', 'SONNBLICK', 'STOCKHOLM', 'VALENTIA']

for station in remaining_stations:
    station_cols = [col for col in weather_cleaned.columns if station in col]
    missing_count = weather_cleaned[station_cols].isnull().sum().sum()
    
    if missing_count > 0:
        print(f"\n{station}: {missing_count} fehlende Werte")
        missing_details = weather_cleaned[station_cols].isnull().sum()
        print(missing_details[missing_details > 0])

# Find observations with many missing values
print("\n\n4. Observations with a lot of missing values:")
print("="*80)

# Count observation types
observation_types = ['cloud_cover', 'wind_speed', 'humidity', 'pressure', 
                     'global_radiation', 'precipitation', 'snow_depth', 
                     'sunshine', 'temp_mean', 'temp_min', 'temp_max']

for obs_type in observation_types:
    obs_cols = [col for col in weather_cleaned.columns if obs_type in col and 'DATE' not in col and 'MONTH' not in col]
    station_count = len(obs_cols)
    print(f"{obs_type:20s}: {station_count:2d} Stationen")

print("\n\nAttention: Observations with less than 15 stations should be dropped.")
```
